```python
colnames=['T','X','Y','Z','VX', 'VY', 'VZ']
colnames_ex = ['T','X','Y','Z','VX', 'VY', 'VZ', 'T', 'U']
file11name = "../.vscode/build/Debug/output/test2/body1.csv"
file12name = "../.vscode/build/Debug/output/test2/body2.csv"
data11 = pd.read_csv(file11name,skiprows=1, sep='\t',names=colnames,header=None)
data12 = pd.read_csv(file12name,skiprows=1, sep='\t',names=colnames,header=None)
print(data11)
```

```python
plt.figure(figsize=(8,8))
plt.title("Tesz: Föld és Nap", fontsize=28)
plt.plot(data11.X,data11.Y, linestyle='--', color='blue')
plt.plot(data12.X,data12.Y, linestyle='-', color='red')
plt.plot(data11.X[-1:], data11.Y[-1:], marker='.', markersize=24, color='orange')
plt.plot(data12.X[-1:], data12.Y[-1:], marker='.', markersize=16, color='red')
plt.xlabel('X koordináta [AU]', fontsize='22')
plt.ylabel('Y koordináta [AU]', fontsize='22')
plt.grid()
plt.savefig('pics/test1.pdf', format='pdf')
plt.show()
```
