```python
#velocity calc for stars
M1 = 1.9e30
M2 = 1.9e30
G = 1.9838 * 1e-29
R1 = 0.05
a = R1
ecc = 0
theta = 0
r = a*(1-ecc**2) / (1-ecc*np.cos(theta))
print(r)
v_stars = (G*(M1 + M2)*(2/R1 - 1/R1)) ** ( 1/2)
print(v_stars/2)

v_stars2 = (G * (M1+M2) / (2*R1) )**(1/2) / 2
print(v_stars2)
```
